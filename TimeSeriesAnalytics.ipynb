```python
import pandas
path = '../COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
all_data = pandas.read_csv(path)
russia_data = all_data[all_data['Country/Region'] == 'Russia']
only_data = russia_data.drop(['Province/State', 'Country/Region', 'Lat', 'Long'], axis=1)
transposed: pandas.DataFrame = only_data.transpose()
diffs = []
prev_value = 0
for index, row in transposed.iterrows():
    value = row[187]
    diff = (value - prev_value)
    diffs.append(diff)
    prev_value = value
transposed['diff'] = diffs
diff_only = transposed[['diff']]
diff_only.plot(figsize=(20,10))
```

```python
# Previous month
diff_only_latest = diff_only[-60:-30]
diff_only_latest.plot(figsize=(20,10))
```

```python
# Current month
diff_only_latest = diff_only[-30:]
diff_only_latest.plot(figsize=(20,10))
```
